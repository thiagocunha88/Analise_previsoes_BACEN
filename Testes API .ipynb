```python
%%time

indices = ['Câmbio', 'INPC', 'IPCA', 'Taxa de desocupação']
qtde_linhas = 630

codigo = 'ExpectativaMercadoMensais'
link_api = f"https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata" \
f"/{codigo}?$top={qtde_linhas}&$skip=0&$filter=baseCalculo%20eq%201&" \
"$format=json&$select=Indicador,Data,DataReferencia,Media,Mediana,Minimo,Maximo,baseCalculo"

# 

#f"https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata" \
#f"/{codigo}?$top={qtde_linhas}&$filter=baseCalculo%20eq%201&" \
#"$format=json&$select=Indicador,Data,DataReferencia,Media,Mediana,Minimo,Maximo,baseCalculo"


consulta_api = pd.read_json(link_api)
indicadores_tabelados = pd.json_normalize(consulta_api.value)
indicadores_tabelados['codigoRecurso'] = codigo
indicadores_tabelados.DataReferencia = [dt.strptime(indicadores_tabelados.DataReferencia[i], '%m/%Y') 
                                        for i in range(len(indicadores_tabelados.DataReferencia))]
indicadores_tabelados.Data = [dt.strptime(indicadores_tabelados.Data[i], '%Y-%m-%d') 
                              for i in range(len(indicadores_tabelados.Data))]
indicadores_tabelados
```

```python
%%time

indices = ['Balança comercial', 'PIB Total', 'IGP-M', 'IPCA',
           'Selic', 'Câmbio', 'Taxa de desocupação']

qtde_linhas = 9999999

# $skip=0&$filter=baseCalculo%20eq%201&

codigo = 'ExpectativasMercadoAnuais'
link_api = f"https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata" \
f"/{codigo}?$top={qtde_linhas}&" \
"$format=json&$select=Indicador,IndicadorDetalhe,Data,DataReferencia,Media,Mediana," \
"DesvioPadrao,Minimo,Maximo,numeroRespondentes,baseCalculo"

#f"https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata" \
#f"/{codigo}?$top={qtde_linhas}&$filter=baseCalculo%20eq%201&" \
#"$format=json&$select=Indicador,Data,DataReferencia,Media,Mediana,Minimo,Maximo,baseCalculo"


consulta_api = pd.read_json(link_api)
indicadores_tabelados = pd.json_normalize(consulta_api.value)
indicadores_tabelados['codigoRecurso'] = codigo
indicadores_tabelados.Data = [dt.strptime(indicadores_tabelados.Data[i], '%Y-%m-%d') 
                              for i in range(len(indicadores_tabelados.Data))]
indicadores_tabelados
```

```python
indices = ['IPCA', 'Selic']

filtro =    (indicadores_tabelados.Indicador.isin(indices)) & \
            (indicadores_tabelados.baseCalculo == 0) & \
            (indicadores_tabelados.Data >= '1998-01-01')

indicadores_tabelados_anuais = indicadores_tabelados[filtro]
indicadores_tabelados_anuais
```
